```python
import  numpy
```

```python
f1 = open('file1.txt', 'r')
file1 = f1.read()
f1.close()
f2 = open('file2.txt', 'r')
file2 = f2.read()
f2.close()
```

```python
def only_names(file):
    testword = ' user'
    userlist = []
    wholelist = file[1:].split(',') #с 1, а не с 0, чтобы убрать " d yfxfkt afqkf
    for i in range (0,len(wholelist)):
        if wholelist[i][0:4] == testword[1:]: 
            userlist.append(wholelist[i][6:])
        elif wholelist[i][0:5] == testword: 
            userlist.append(wholelist[i][7:])
        else: pass
    
    return set(userlist)
```

```python
userlist1 = only_names(file1)
userlist2 = only_names(file2)
```

```python
def final_filter(userlist1,userlist2):
    final_list = []
    for item1 in userlist1:
        flag = True
        for item2 in userlist2:
            if item1 == item2: flag = False
            else: pass
        if flag == True: final_list.append(item1)
    return final_list

userlist = final_filter(userlist1,userlist2)
print(len(userlist))
```

```python
for item in userlist: print(item)
```
